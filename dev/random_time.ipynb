```python
"""
Give time to Fireloc Contributions
"""

import datetime
import random
from calendar import monthrange
from gasp3.fm import tbl_to_obj
from gasp3.gt.to.shp import df_to_shp

START_DATE = datetime.datetime(2019, 9, 13, 19, 0, 0)
END_DATE   = datetime.datetime(2019, 9, 14, 13, 30, 0)
SAMPLE     = r'F:\fireloc\tst_miranda\pnt_obs_sample.shp'
OUT_SAMPLE = r'F:\fireloc\tst_miranda\pnt_obs_sample_date.shp'
```

```python
gDf = tbl_to_obj(SAMPLE)
```

```python
# Get Random Dates

def get_year(start, end):
    # Get Year
    if start.year == end.year:
        year = start.year
    
    else:
        year = random.randint(start.year, end.year)
    
    return year

def get_month(start, end, year):
    # Get Month
    if start.year == end.year:
        if start.month == end.month:
            month = start.month
        else:
            month = random.randint(start.month, end.month)
    
    else:
        if year == start.year:
            month = random.randint(start.month, 12) if start.month < 12 else 12
        
        elif year == end.year:
            month = random.randint(1, end.month) if end.month > 1 else 1
    
    return month


def get_day(s, e, y, m):
    # Get Day
    ndays = monthrange(y, m)[1]
    
    if s.year == e.year and s.month == e.month:
        if s.day == e.day:
            day = s.day
        else:
            day = random.randint(s.day, e.day)
    
    elif s.year == e.year and s.month != e.month:
        
        if m == s.month:
            day = random.randint(m, ndays)
        elif m == e.month:
            day = random.randint(1, e.day)
        else:
            day = random.randint(1, ndays)
    
    elif s.year != e.year:
        if y == s.year:
            if m == s.month:
                day = random.randint(m, ndays)
            else:
                day = random.randint(1, ndays)
        elif y == e.year:
            if m == e.month:
                day = random.randint(1, e.day)
            else:
                day = random.randint(1, ndays)
        else:
            day = random.randint(1, ndays)
    
    return day


def get_hour(s, e, y, m, d):
    # Get Hour
    sDay = datetime.datetime(s.year, s.month, s.day)
    eDay = datetime.datetime(e.year, e.month, e.day)
    cDay = datetime.datetime(y, m, d)
    
    if sDay == eDay:
        hour = random.randint(s.hour, e.hour)
    else:
        if sDay == cDay:
            hour = random.randint(s.hour, 23)
        elif eDay == cDay:
            hour = random.randint(0, e.hour)
        else:
            hour = random.randint(0, 23)
    
    return hour


def get_minute(s, e, y, m, d, h):
    # Get minute
    
    sHour = datetime.datetime(s.year, s.month, s.day, s.hour)
    eHour = datetime.datetime(e.year, e.month, e.day, e.hour)
    cHour = datetime.datetime(y, m, d, h)
    
    if sHour == eHour:
        minute = random.randint(s.minute, e.minute)
    else:
        if sHour == cHour:
            minute = random.randint(s.minute, 59)
        elif eHour == cHour:
            minute = random.randint(0, e.minute)
        else:
            minute = random.randint(0, 59)
    
    return minute

def get_second(s, e, y, m, d, h, mi):
    # Get second
    
    sMinute = datetime.datetime(s.year, s.month, s.day, s.hour, s.minute)
    eMinute = datetime.datetime(e.year, e.month, e.day, e.hour, e.minute)
    cMinute = datetime.datetime(y, m, d, h, mi)
    
    if sMinute == eMinute:
        second = random.randint(s.second, e.second)
    else:
        if sMinute == cMinute:
            second = random.randint(s.second, 59)
        elif eMinute == cMinute:
            second = random.randint(0, e.second)
        else:
            second = random.randint(0, 59)
    return second

def sanitize(s):
    return "0{}".format(str(s)) if len(str(s)) == 1 else str(s)

DATES = []
TIMES = []
for i in range(gDf.shape[0]):
    year = get_year(START_DATE, END_DATE)
    
    month = get_month(START_DATE, END_DATE, year)
    
    day = get_day(START_DATE, END_DATE, year, month)
    
    hour = get_hour(START_DATE, END_DATE, year, month, day)
    
    minute = get_minute(START_DATE, END_DATE, year, month, day, hour)
    
    second = get_second(START_DATE, END_DATE, year, month, day, hour, minute)
    
    month, day, hour, minute, second = [sanitize(i) for i in [month, day, hour, minute, second]]
    
    DATES.append('{}-{}-{}'.format(year, month, day))
    TIMES.append('{}:{}:{}'.format(hour, minute, second))
```

```python
gDf['date'] = DATES; gDf['time'] = TIMES
```

```python
# Export
df_to_shp(gDf, OUT_SAMPLE)
```
